```python
df = pd.read_csv("econ.obs.csv")
df = df.rename(columns={"rebased_share_index": "value"})
df = df[['date', 'value', 'party', 'group']]

df['date'] = pd.to_datetime(df['date'], dayfirst=True)

df

base = alt.Chart(df).encode(
    x=alt.X('date:T', title=""),
    y=alt.Y('value:Q', title="Share index", 
            axis=alt.Axis(titleY=-17),
            scale=alt.Scale(domain=[80, 140])),
    color=alt.Color('party:N', title="Party", 
                    legend=alt.Legend(title="", orient="none", labelFontSize=12, titleFontSize=14, direction="horizontal", legendY=-15),
                    scale=alt.Scale(domain=['Conservative', 'Labour', 'Coalition'], range=['#0087DC', 'hsl(2, 75.30%, 49.20%)', 'rgb(199, 35, 214)'])),
    detail='group:N'
)

lines = base.mark_line()

chart = lines

chart = chart.properties(
    width=450,
    height=300
)

chart.save("fig1.png", scale_factor=3.0)
chart.save("fig1.json")
chart
```
