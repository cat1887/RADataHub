```python
import pandas as pd
import altair as alt
import numpy as np
import eco_style
import json

alt.themes.enable('light')
```

# Fig 1

```python
df = pd.read_excel("cleaned.xlsx", sheet_name="fig1")

height = 300
width = 500

df['label'] = np.where(df['date'] == '2025-10-01', df['series'], '')
df['value'] = df['value'] / 100

base = alt.Chart(df).encode(
    x=alt.X("date:T"),
    y=alt.Y("value:Q", title="", axis=alt.Axis(format='%')),
    color=alt.Color("series:N", legend=None, scale=alt.Scale(domain=['Wales', 'Scotland', 'N. Ireland', 'UK'], range=["#36B7B4","hsl(347, 79.70%, 51.80%)","#F4C245", 'hsl(347, 0%, 40.80%)'])),
)

dy_offsets = {
    "Wales": 0,
    "Scotland": 0,
    "N. Ireland": -10,
    "England": 0,
}

lines = base.mark_line().encode(
    strokeDash=alt.condition(
        alt.datum.series == 'UK',
        alt.value([3, 3]),
        alt.value([0])
    ),
)

end_labels = base.mark_text(align='left',
                             dy=alt.expr(json.dumps(dy_offsets).replace('"', "'")+ '[datum.series]'),
                             dx=5).encode(
    text='label'
)

zero_line = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='black', size=0.5).encode(y='y')

# shade 2024-04-01 onwards as a forecast
forecast = alt.Chart(pd.DataFrame({'start': ['2024-04-01'], 'end': ['2025-10-01']})).mark_rect(opacity=0.05).encode(
    x='start:T',
    x2='end:T',
    color=alt.value('black'),
    y=alt.value(0),
    y2=alt.value(height)
)

forecase_label = alt.Chart(pd.DataFrame({'x': ['2024-04-01'], 'y': [0.5], 'text': ['Forecast']})).mark_text(
    align='left',
    opacity=0.8
).encode(
    x='x:T',
    y=alt.value(-5),
    text='text'
)

forecast_rule = alt.Chart(pd.DataFrame({'x': ['2024-04-01']})).mark_rule(color='black', opacity=0.35).encode(x='x:T')


chart = (forecast+ forecast_rule+  lines + end_labels + zero_line + forecase_label)

chart = chart.properties(
    width=width,
    height=height,
)

chart.save("fig1.json")
chart.save("fig1.png", scale_factor=2.0)

chart
```

```python
df = pd.read_excel("cleaned.xlsx", sheet_name="fig3")
df['series'] = np.where(df['series'] == 'UK avg.', 'UK', df['series'])


height = 300
width = 500

charts = []
for i, series in enumerate(["Wales", "Scotland", "N. Ireland", "UK"]):
    temp = df[df['series'] == series]

    base = alt.Chart(temp).encode(
        x=alt.X("date_label:N", title="", axis=alt.Axis( ticks=False, labelAngle=0, labelFontSize=11, labelColor="black")),
        y=alt.Y("value:Q", title="", axis=alt.Axis(labelExpr="'£' + datum.label") if i == 0 else None),
        color=alt.Color("date_label:N", scale=alt.Scale(), legend=None),
    )

    bars = base.mark_bar()

    rule = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='black', size=1.5).encode(
        y=alt.datum(35.32),
        x=alt.value(0),
        x2=alt.value((width/4)+5),
        strokeDash=alt.value([3, 3])
    )

    rule_label = alt.Chart(pd.DataFrame({'x': [0], 'y': [35.32], 'text': ['2025Q4 Avg.']})).mark_text(
        align='left',
        fontSize=12,
        opacity=0.8
    ).encode(
        y=alt.datum(35.32),
        x=alt.value(width/4+5),
        text='text' if i == 3 else alt.value('')    
    )

    chart = bars + rule + rule_label

    chart = chart.properties(
        title=alt.TitleParams(text=series,
                              fontWeight='normal',
                              dy=5,
                               anchor='middle', orient='bottom'),
        width=width/4,
    )

    charts.append(chart)

chart = alt.hconcat(*charts, spacing=20)

chart.save("fig3.json")
chart.save("fig3.png", scale_factor=2.0)

chart
```
