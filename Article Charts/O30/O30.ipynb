```python
import pandas as pd
import numpy as np
import altair as alt
import eco_style

alt.themes.enable('light')
```

```python
countries = ["SSD","ZWE","SWZ","MDG","GNB","GTM","BDI","COD","STP","HTI","ZAF","TGO","ZMB","GIN","LSO","AFG","GMB","MWI","HND","SLE"]
len( countries)
```

```python
df = pd.read_csv('O30.csv')
df = df[[
    "REF_AREA:Geographic area",
    "OBS_VALUE:Observation Value"
]]

df.columns = ['geo', 'value']
df["iso3"] = df['geo'].str.split(':').str[0]
df["country"] = df['geo'].str.split(':').str[1]
df['value'] = df['value'].astype(float)/100

rebinds = {
    'COD' : "DR Congo",
}
df['country'] = df.apply(lambda x: rebinds.get(x['iso3'], x['country']), axis=1)

df = df[df['iso3'].isin(countries)]

chart = alt.Chart(df).mark_bar(
    color="#36B7B4"
).encode(
    y=alt.Y('country:N', sort='-x', title=''),
    x=alt.X('value:Q', axis=alt.Axis(format='%')),
)

chart.save("O30_fig1.json")
chart.save("O30_fig1.png", scale_factor=2.0)

chart
```
