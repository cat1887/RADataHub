```python
import pandas as pd
import altair as alt
import numpy as np
import os
import eco_style

alt.themes.enable('light')
```

### Fig 1

```python
df = pd.read_csv("fig1_wheat_imports_change.csv")


chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('source:N', axis=alt.Axis(domain=False, tickOpacity=0, labels=False)),
    y=alt.Y('value:Q', axis=alt.Axis(title="Tonnes", titleY=-20, labelExpr="datum.value + 'M'")),
    color=alt.Color('source:N', legend=alt.Legend(orient="none", title=None, legendY=-16, legendX=0, direction="horizontal")),
    column=alt.Column('country:N', header=alt.Header(orient='bottom', title=""), sort=alt.EncodingSortField(field="Country", order="descending")),
).properties(
    width=50,
    height=300
)

chart.save("fig1.json")
chart.save("fig1.png", scale_factor=2.0)

chart.display()
```

## Fig 2

```python
df = pd.read_csv('fig2_total_wheat_exports.csv')

base = alt.Chart(df).encode(
    x=alt.X('year:O', axis=alt.Axis()),
    y=alt.Y('value:Q', axis=alt.Axis(title="Tonnes", labelExpr="datum.value + 'M'")),
    color=alt.Color('country:N', legend=None)
)

lines = base.mark_line()

labels = base.transform_filter("datum.year == 2024").mark_text(
    align='left',
    baseline='middle',
    dx=5
).encode(
    text='country:N'
)

chart = lines  + labels

chart.display()

chart.save("fig2.json")
chart.save("fig2.png", scale_factor=2.0)
```

## Fig 3

```python
df = pd.read_csv("fig3_grain_prices.csv")

df = df.melt(id_vars=["Month"], value_name="value", var_name="series")

df['date'] = pd.to_datetime(df['Month'], format='%YM%m')

base = alt.Chart(df).encode(
    x=alt.X('date:T', axis=alt.Axis()),
    y=alt.Y('value:Q', axis=alt.Axis(title=None)),
    color=alt.Color('series:N', legend=None)
)

lines = base.mark_line()

labels = base.transform_filter("datum.Month == '2023M12'").mark_text(
    align='left',
    baseline='middle',
    dx=5,
    dy=alt.expr("datum.series == 'Grain' ? -5 : 0")
).encode(
    text='series:N'
)

chart = lines  + labels

chart.save("fig3.json")
chart.save("fig3.png", scale_factor=2.0)

chart.display()
```

## Fig 4

```python
df = pd.read_csv("fig4_changes.csv")

df['value'] = df.Change.str.replace("%", "").astype(float)/100

countries = [
    "Liberia",
    "Zimbabwe",
    "Argentina",
    "Angola",
    "Lesotho",
    "Niger",
    "Nigeria",
    "Swaziland",
    "Ethiopia",
    "US",
    "UK",
    "EU"
]

df = df[df.Country.isin(countries)]

base = alt.Chart(df).encode(
    x=alt.X('value:Q', axis=alt.Axis(title=None, format=".0%"), scale=alt.Scale(domainMin=-0.3)),
    y=alt.Y('Country:N', axis=alt.Axis(title=None), sort="-x"),
).properties(
    width=300,
    height=300
)

bars = base.mark_bar(color="#36B7B4")

labels = base.mark_text(
    align='left',
    baseline='middle',
    dx=alt.expr("datum.value > 0 ? 5 : -25")
).encode(
    text='Change:N'
)

chart = bars + labels

chart.save("fig4.json")
chart.save("fig4.png", scale_factor=2.0)

chart.display()
```

# COTW

```python
df = pd.read_json("https://eco-temp-cache.s3.eu-west-2.amazonaws.com/ukr/gexpo.json").query("value < 1.815136e+09")

df['value'] = df['value'] / 1000

base = alt.Chart(df).encode(
    x=alt.X('date:T', axis=alt.Axis(title=None)),
    y=alt.Y('value:Q', axis=alt.Axis(title="Tonnes", format=".0s")),
)

lines = base.mark_line(interpolate='monotone')

chart = lines


chart = chart.properties(
    title=alt.TitleParams(
        "Ukraine's Grain Exports",
        subtitle=["Monthly grain exports from Ukraine, 2019-2023", "Data: COMTRADE", ""],
        anchor="start",
        color="#676A86",
        subtitleColor="#676A86",
        dx=30,
    )
)

chart.save("COTW_UKR.json")
chart.save("COTW_UKR.png", scale_factor=2.0)


chart
```
