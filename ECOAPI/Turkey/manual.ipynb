#### Inflation

```python
infl_raw = pd.read_excel('raw_data/index numbers and rate of changes in the consumer price index.xls', header=None)
infl_raw.head()
```

Find header & footer index

```python
header_idx = infl_raw[infl_raw[0].eq('Consumer price index numbers (2003=100)')].index.values[0] + 2

# Find footer index, 2nd NaN value after header index
# starting from header index (months row), sum NaN values until 2nd NaN value, then find max index, then subtract 1
footer_idx = infl_raw[header_idx:].isna().cumsum().eq(2).idxmax().values[0] - 1
```

```python
# specify number of rows to read after header row
num_rows = footer_idx - header_idx
df_infl_wide = pd.read_excel(
    io='raw_data/index numbers and rate of changes in the consumer price index.xls', 
    header=header_idx,
    nrows=num_rows,
    usecols="A:M", 
    index_col=0
)
```

```python
# pivot from wide to long-form data table
df_infl = df_infl_wide.stack().reset_index()

# combine text in columns 0 and 1
df_infl['level_0'] = df_infl['level_1'].astype(str) + ' ' + df_infl['level_0'].astype(str)
df_infl = df_infl.drop('level_1', axis=1)
df_infl.columns = ['date', 'value']

df_infl['date'] = pd.to_datetime(df_infl['date'], format='%B %Y')
```

#### Growth

```python
dataset = 'GDPbyEx.xls'
raw_data = pd.read_excel('raw_data/' + dataset, header=None)
```

```python
raw_data.head(10)
```

```python
header_idx = raw_data[raw_data[2].eq('Yıllık\nAnnual')].index.values[0] + 1
# Find footer index, 3rd NaN value after header index in year column
footer_idx = raw_data.iloc[header_idx:, 1].isna().cumsum().eq(3).idxmax() - 1
```

```python
# specify number of rows to read after header row
num_rows = footer_idx - header_idx

# `usecols`: use column B (for Year) and columns J to M (for values)
df_grow_wide = pd.read_excel(
    io='raw_data/' + dataset, 
    header=header_idx,
    nrows=num_rows,
    usecols="B,J:M",
    index_col=0
)
```

```python
# try and convert row values to numeric, if not possible, set to NaN
df_grow_wide = df_grow_wide.apply(pd.to_numeric, errors='coerce')

# drop if all values are NaN (i.e. no data for entire row / year)
df_grow_wide = df_grow_wide.dropna(how='all')
df_grow_wide.info()
```

```python
# pivot from wide to long-form data table
df = df_grow_wide.stack().reset_index()

# in `level_0` column, remove '(e)' string from year values
df['level_0'] = df['level_0'].astype(str).str.replace('(e)', '')

# Convert `level_1` column values from Roman numerals to Quarters
df['level_1'] = df['level_1'].map(quarter_map)

# combine text in columns 0 and 1
df['level_0'] = df['level_0'].astype(str) + '-' + df['level_1'].astype(str)
df = df.drop('level_1', axis=1)
df.columns = ['date', 'value']
```

```python
df.tail(30)
```

Regular Expression (Q\d) (\d+):

- (Q\d) captures a letter 'Q' followed by a digit (quarter) and stores it in a group.
- (\d+) captures one or more digits (year) and stores it in a group.

Replacement Pattern r'\2-\1':
- \2 refers to the second captured group, which is the year.
- \1 refers to the first captured group, which is the quarter.

```python
df['date'] = pd.to_datetime(df['date'].str.replace(r'(Q\d) (\d+)', r'\2-\1'), errors='coerce')
df['date'] = df['date'].astype(str)
```

```python
# Export annualised growth (by quarter) to CSV
df.to_csv('data/grow.csv', index=False)

# Export df to JSON, array format with objects as elements with two key-value pairs (date and value)
df.to_json('data/grow.json', orient='records')
```

#### Inequality

Income Distribution Stats: https://data.tuik.gov.tr/Bulten/Index?p=Income-Distribution-Statistics-2022-49745

```python
dataset = 'gini coefficient s80_s20 and s90_s10 ratios.xls'
raw_data = pd.read_excel('raw_data/' + dataset, header=None)

raw_data.head(10)
```

```python
header_idx = raw_data[raw_data[1].eq('Yıllar - Years')].index.values[0] + 1
```

```python
footer_idx = raw_data[raw_data[0].eq('Gini katsayısı - Gini coefficient')].index.values[0]
```

```python
df_wide = pd.read_excel(
    io='raw_data/' + dataset, 
    header=header_idx,
    nrows=footer_idx - header_idx,
    index_col=0
)
```

```python
df = df_wide.stack().reset_index(level=1).reset_index(drop=True)
df.columns = ['date', 'value']
df['date'] = pd.to_datetime(df['date'], format='%Y')
df['date'] = df['date'].astype(str)
# convert 'value' column to unformatted percentage (i.e. 100 scale)
df['value'] = df['value'] * 100
df['value'] = df['value'].round(2)
```

```python
# export as csv and json
df.to_csv('data/ineq.csv', index=False)

df.to_json('data/ineq.json', orient='records')
```
