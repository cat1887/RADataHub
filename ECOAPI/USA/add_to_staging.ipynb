```python
staging = pd.read_csv('/Users/joshhellings/Documents/ECO/api/metadata-staging.csv')

# eco/{}/{}
eco = 'https://api.economicsobservatory.com'

localAPI = 'https://api.stlouisfed.org/fred/series/observations?api_key=22ee7a76e736e32f54f5df0a7171538d&file_type=json&series_id={}'

series = ['popu', 'empl', 'empd', 'part', 'gdp', 'gdpp']

fred = ['POPTHM', 'LREM64TTUSM156S', 'CE16OV', 'CIVPART', 'GDP', 'A939RC0Q052SBEA']

# create 'USA' df
iso = 'usa'
stitle_suffix = '| FRED through ECO API'
num_series = len(series)

usa_data = pd.DataFrame({
    'country_full': ['USA']*num_series,
    'country': [f'{iso}']*num_series,
    'ecoSeries': series,
    'ecoAPI':[
        f'{eco}/{iso}/{series[0]}',
        f'{eco}/{iso}/{series[1]}',
        f'{eco}/{iso}/{series[2]}',
        f'{eco}/{iso}/{series[3]}',
        f'{eco}/{iso}/{series[4]}',
        f'{eco}/{iso}/{series[5]}'
    ],
    'localAPI':[
        localAPI.format(fred[0]),
        localAPI.format(fred[1]),
        localAPI.format(fred[2]),
        localAPI.format(fred[3]),
        localAPI.format(fred[4]),
        localAPI.format(fred[5])
    ],
    # list of FRED series IDs
    'localSeries': fred,
    'localDataset': [None]*num_series,
    'Title': [
        'Population',
        'Employment Rate',
        'Employed',
        'Participation Rate',
        'GDP',
        'GDP per capita'
    ],
    'SubTitle': [
        f'Population {stitle_suffix}',
        f'All persons, aged 15-64, % {stitle_suffix}',
        f'Employment level, thousands {stitle_suffix}',
        f'Labour force participation, % {stitle_suffix}',
        f'Current prices, $bn {stitle_suffix}',
        f'Current prices, $bn {stitle_suffix}'
    ],
    'API_base': [
        'https://api.stlouisfed.org/fred/series/observations?api_key=22ee7a76e736e32f54f5df0a7171538d&file_type=json&series_id='
        ]*num_series,
    'API_full': [
        localAPI.format(fred[0]),
        localAPI.format(fred[1]),
        localAPI.format(fred[2]),
        localAPI.format(fred[3]),
        localAPI.format(fred[4]),
        localAPI.format(fred[5])
    ],
    'source': ['FRED']*num_series,
    'source_long': ['US Federal Reserve Economic Data']*num_series,
    'description': [
        'Updated monthly',
        'Updated monthly',
        'Updated monthly',
        'Updated monthly',
        'Updated quarterly',
        'Updated quarterly'
    ],
    'units': [
        'Thousands, Not Seasonally Adjusted',
        'Percent, Seasonally Adjusted',
        'Thousands of Persons, Seasonally Adjusted',
        'Percent, Seasonally Adjusted',
        'US$ billion, Seasonally Adjusted Annual Rate',
        'US$, Seasonally Adjusted Annual Rate'
    ],
    'date': ['observations.date']*num_series,
    'field': ['observations.value']*num_series,
    'multiplier': [1]*num_series,
    'date_keys': [None]*num_series,
    'value_keys': [None]*num_series,
    'filter': [None]*num_series,
    'orient': [None]*num_series,
    'use_cache': [0]*num_series
})
```

```python
# check if `staging` is alphabetically sorted `country_full`
staging['country_full'].sort_values().unique()
```

```python
# add `usa_data` to `staging`
staging = pd.concat([staging, usa_data], ignore_index=True)
```

```python
# sort `staging` alphabetically by `country_full`
staging = staging.sort_values('country_full')
```
