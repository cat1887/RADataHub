```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ch5_x_assets_liabilities.csv")
df["date"] = df["date"].str.replace(":", " ").str.replace("Q1", "Jan").str.replace("Q2", "Apr").str.replace("Q3", "Jul").str.replace("Q4", "Oct")
df = df.rename(columns={"FL152090005.Q": "Household Net Worth", "FL152000005.Q": "Household Assets", "FL154190005.Q": "Household Liabilities"})
```

```python
df = df.melt(id_vars="date")
```

```python
df["year"] = df["date"].str.split(" ").str[0].astype(int).reset_index(drop=True)
df = df[df["year"]>=2001]
```

```python
df
```

```python

df["value"] = df.apply(lambda x: x["value"]*-1 if x["variable"]=="Household Liabilities" else x["value"], axis=1)
```

```python
df.melt()
```

```python
df.to_csv("ch5_x_assets_liabilitiesv6.csv")
```

```python
df1 = df[~df["variable"].str.contains("Household Net Worth")].reset_index(drop=True).rename(columns={"variable": "name"})
df2 = df[df["variable"].str.contains("Household Net Worth")].reset_index(drop=True)
```

```python
df1.to_csv("ch5_x_assets_liabilities_pt1.csv")
df2.to_csv("ch5_x_assets_liabilities_pt2.csv")
```

```python
df = pd.read_csv("Ass_laibilities_data.csv")
df["date"] = df["date"].str.replace(":", " ").str.replace("Q1", "Jan").str.replace("Q2", "Apr").str.replace("Q3", "Jul").str.replace("Q4", "Oct")
df = df.rename(columns={"FL153165105.Q": "Home Mortgages", "FL154199005.Q": "Other Liabilities", "FL153166000.Q": "Consumer Credit", "FL154090005.Q": "Financial Assets", "LM152010005.Q": "Non-Financial Assets"})
df["year"] = df["date"].str.split(" ").str[0].astype(int).reset_index(drop=True)
df = df[df["year"]>=2001]
df = df.drop("year", axis=1)
```

```python
df
```

```python
df1 = df[["date", "Financial Assets", "Non-Financial Assets"]]
df2 = df[["date","Consumer Credit", "Other Liabilities", "Home Mortgages"]]
```

```python
df_assets = df1.melt(id_vars="date", var_name="Assets", value_name="asset_val")
df_liab = df2.melt(id_vars="date", var_name="Liabilities", value_name="liabil_val")
df3 = pd.concat([df_liab, df_assets], axis=0)
df3
```
