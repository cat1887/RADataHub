```python
df = pd.read_csv('lad_prod.csv')

# get the median for each date
df['median'] = df.groupby('date')['value'].transform('median')
df['relative'] = df['value'] / df['median']

df.relative.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9])
```

```python

print(df.relative.describe())

df = df.drop(columns=['median'])
df['value'] = df['relative']

df = df.pivot_table(index=['area_code', 'name'], columns='date', values='value')
df = df.reset_index()

df.to_csv('lad_prod_rel_wide.csv', index=False)
```
