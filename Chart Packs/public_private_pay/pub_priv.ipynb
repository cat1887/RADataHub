```python
import pandas as pd
import altair as alt
import numpy as np
import eco_style
alt.themes.enable('light')
```

```python
df = pd.read_csv("ifs_pub_priv.csv")
df['date'] = pd.to_datetime(df['monthyear'], format='%b %Y')
df = df[['date', 'Private', 'Public']]
df = df.query('date >= "2010-01-01"')
start = df.iloc[0]
df['Private'] = df['Private'] / start['Private']
df['Public'] = df['Public'] / start['Public']
df = df.melt('date', var_name='series', value_name='value')
df['value'] = df['value']-1

df['label'] = np.where(df['date'] == '2023-11-01', df['series'], '')


base = alt.Chart(df).encode(
    x=alt.X('date:T', axis=alt.Axis(title='')),
    y=alt.Y('value:Q', axis=alt.Axis(title='', format="%"), scale=alt.Scale(domain=[-0.1, 0.1])),
    color=alt.Color('series:N', legend=None),
)

labels = base.mark_text(align='left', dx=5, dy=0).encode(
    text='label:N'
)

line = base.mark_line()

chart = line + labels

chart = chart.properties(
    width=450,
    height=300,
    title=alt.TitleParams(
        'Pay - public and private',
        subtitle=["Change in mean real weekly earnings since Jan 2010", "Source: IFS, own calculations", ""],
        anchor='start',
        #fontSize=18,
        #subtitleFontSize=14
    )
)

chart.save("pub_priv.json")
chart.save("pub_priv.png", scale_factor=3)
chart
```
