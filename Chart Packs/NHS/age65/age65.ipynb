```python
import pandas as pd
import numpy as np
import altair as alt
import eco_style
alt.themes.enable('light')
```

```python
df = pd.read_csv('requests_for_support.csv')
df['Series'] = df.Series.str.strip()
# Region	Value	Series

df = df[['Region', 'Value', 'Series']]
df = df.pivot_table(index='Region', columns='Series', values='Value').reset_index()
df['Total Number of Requests'] = df['Total Number of Requests']-df['No Services Provided - Deceased']-df['No Services Provided']
df = df.melt(id_vars='Region', value_vars=['Total Number of Requests', 'No Services Provided', 'No Services Provided - Deceased'], var_name='Series', value_name='Value')
df

# replace Yorkshire and The Humber with Yorksh. & Humber
region_maps = {
    'East Midlands': 'E. Midlands',
    'West Midlands': 'W. Midlands',
    'East of England': 'E. of England',
    'North East': 'N. East',
    'North West': 'N. West',
    'South East': 'S. East',
    'South West': 'S. West',
    'Yorkshire and The Humber': 'Yorksh. & Humber'
}

df['Region'] = df['Region'].replace(region_maps)

df
```

```python


bars = alt.Chart(df.query("Series != 'Total Number of Requests'")).mark_bar().encode(
    x=alt.X("Region:N"),
    y=alt.Y("Value:Q"),
    color=alt.Color("Series:N", 
                    
                    scale=alt.Scale(
                        domain=['No Services Provided', 'No Services Provided - Deceased' , 'Total Number of Requests'],
                        range=['#36B7B4', '#E6224B', 'black']),
                    legend=alt.Legend(
                        labelLimit=2000000,
        orient="top",
        values = ['No Services Provided', 'No Services Provided - Deceased', 'Total Number of Requests'],
          title=None)),
)

points = alt.Chart(df.query("Series == 'Total Number of Requests'")).mark_circle(size=5).encode(
    x=alt.X("Region:N"),
    y=alt.Y("Value:Q", title=''),
    color=alt.value('black'),
    size=alt.value(100),
)

chart = (bars + points).properties(
    width=625,
    height=300
)

chart.save('chart.png', scale_factor=3)

chart
```

## Proportions

```python
df = pd.read_csv('requests_for_support.csv')
df['Series'] = df.Series.str.strip()
# Region	Value	Series

df = df[['Region', 'Value', 'Series']]
df = df.pivot_table(index='Region', columns='Series', values='Value').reset_index()
df['Total Number of Requests'] = df['Total Number of Requests']-df['No Services Provided - Deceased']-df['No Services Provided']
df['Value'] = (df['No Services Provided - Deceased'] + df['No Services Provided'])/df['Total Number of Requests']

df = df[['Region', 'Value']]

region_maps = {
    'East Midlands': 'E. Midlands',
    'West Midlands': 'W. Midlands',
    'East of England': 'E. of England',
    'North East': 'N. East',
    'North West': 'N. West',
    'South East': 'S. East',
    'South West': 'S. West',
    'Yorkshire and The Humber': 'Yorksh. & Humber'
}

df['Region'] = df['Region'].replace(region_maps)

df

bars = alt.Chart(df).mark_bar(color="#36B7B4").encode(
    y=alt.Y("Region:N", sort='-x', title=''),
    x=alt.X("Value:Q", axis=alt.Axis(format='%')),
)

chart = (bars).properties(
    width=500,
    height=300
)

chart.save('no_services_proportion.png', scale_factor=3)
chart.save('no_services_proportion.json')
chart

```
