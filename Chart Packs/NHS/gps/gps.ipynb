```python
import pandas as pd
import numpy as np
import altair as alt
import eco_style
import json
alt.themes.enable('light')
```

```python
with open("vals.js") as f:
    vals = f.read()

vals = vals.replace("new Date(", "").replace("0)", "0")

data = json.loads(vals)
data

parsed_data = []

# All GPs HC", "All GPs FTE", "Qualified & permanent FTE", "Qualified & permanent HC

for row in data['data']:
    # 1441065600000
    parsed_row = {}
    parsed_row['date'] = pd.to_datetime(row['label'], unit='ms')
    parsed_row['All GPs HC'] = row['value'][0]
    parsed_row['All GPs FTE'] = row['value'][1]
    parsed_row['Qualified & permanent FTE'] = row['value'][2]
    parsed_row['Qualified & permanent HC'] = row['value'][3]
    parsed_data.append(parsed_row)


df = pd.DataFrame(parsed_data)
df = df.melt('date', var_name='series', value_name='value')
df = df.dropna()
df

df['label'] = np.where(df.date == "2024-01-01", df.series, '')

base = alt.Chart(df).encode(
    x=alt.X('date:T', axis=alt.Axis()),
    y=alt.Y('value:Q', axis=alt.Axis(title=None)),
    color=alt.Color('series:N', legend=None),
)

lines = base.mark_line()

end_label = base.mark_text(align='left', dx=5,     dy=alt.expr("datum.series == 'Qualified & permanent HC' ? 5 : 0")
).encode(
    text='label:N',
)


chart = lines  + end_label

chart = chart.properties(
    width=500,
    height=300
)
chart.save('gps.json')
chart.save('gps.png', scale_factor=3.0)
chart
```
