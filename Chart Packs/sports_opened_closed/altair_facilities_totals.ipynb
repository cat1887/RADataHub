```python
# import facilities data
import pandas as pd
running_totals = pd.read_csv('sports_facilities_opened_closed.csv')
```

```python
# Convert the 'month' column to datetime
running_totals['month'] = pd.to_datetime(running_totals['month'])
```

```python
# Set theme
alt.themes.enable('dark')

# Create the base chart
base = alt.Chart(running_totals).encode(
    x=alt.X('month:T', title='Month')
)

# Create the area chart for closures
closures_area = base.mark_area(opacity=0.7).encode(
    y=alt.Y('closures:Q', title='Total Facilities'),
    color=alt.value('#E6224B'),  # First color from your category scheme
    tooltip=[
        alt.Tooltip('month:T', title='Month', format='%B %Y'),
        alt.Tooltip('closures:Q', title='Facilities Closed', format=',')
    ]
)

# Create the area chart for openings
openings_area = base.mark_area(opacity=0.7).encode(
    y=alt.Y('openings:Q', title='Total Facilities'),
    color=alt.value('#36B7B4'),  # Second color from your category scheme
    tooltip=[
        alt.Tooltip('month:T', title='Month', format='%B %Y'),
        alt.Tooltip('openings:Q', title='Facilities Opened', format=',')
    ]
)

# Calculate final values for annotations
final_values = running_totals[['month', 'openings', 'closures']].iloc[-1]

# Create text annotations for closures
closures_text = alt.Chart(pd.DataFrame({
    'month': [final_values['month']],
    'count': [final_values['closures']],
    'Type': ['Closures']
})).mark_text(
    align='left',
    dx=10,
    dy=-5,
    color='#E6224B'
).encode(
    x='month:T',
    y='count:Q',
    text=alt.Text('count:Q', format=',')
).transform_calculate(
    text='" " + format(datum.count, ",") + " closed"'
).encode(
    text='text:N'
)

# Create text annotations for openings
openings_text = alt.Chart(pd.DataFrame({
    'month': [final_values['month']],
    'count': [final_values['openings']],
    'Type': ['Openings']
})).mark_text(
    align='left',
    dx=10,
    dy=-5,
    color='#36B7B4'
).encode(
    x='month:T',
    y='count:Q',
    text=alt.Text('count:Q', format=',')
).transform_calculate(
    text='" " + format(datum.count, ",") + " opened"'
).encode(
    text='text:N'
)

# Combine the area charts and text annotations
chart = (closures_area + openings_area + closures_text + openings_text).properties(
    width=400,
    height=300,
    title='England: Sports Facilities Opened and Closed Since 2010'
)

# Configure the y-axis
chart = chart.configure_axisY(
    titleAngle=270,  # This rotates the text 270 degrees
    titleAlign='right',
    titleBaseline='middle',
    titleX=-35,
    titleY=110,
    format='~s'
)

# Configure the x-axis
chart = chart.configure_axisX(
    titleBaseline='bottom',
    title='Month',
    titleAlign='center',
    titleY=35,
)

# Display the chart
chart.save('england_facilities_open_close.json')
chart
```
