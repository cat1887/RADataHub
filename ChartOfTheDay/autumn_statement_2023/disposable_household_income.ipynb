```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_excel("real_disposable_household_income_per_person.xlsx")
```

```python
df = df[df.Date <= '2024-09-30']

base = alt.Chart(df).encode(
    x=alt.X('Date:T',         axis=alt.Axis(
            title=None,
            gridColor="#b4c8d8",
            gridDash=[1, 5],
            labelFont="Circular Std",
            gridOpacity=0,
            domainOpacity=0.5,
            labelColor="#b4c8d8",
            labelOpacity=1,
            labelPadding=5,
            domainColor='#b4c8d8',
            tickColor='#b4c8d8'),),
    y=alt.Y('Growth:Q',

            axis=alt.Axis(
                title=None,
                labelColor="#b4c8d8",
                labelFont="Circular Std",
                gridColor="#b4c8d8",
                gridDash=[1, 5],
                gridOpacity=0.5,
                domainOpacity=0.5,


                format="%",
                labelOpacity=1,
                labelFontSize=12,
                labelPadding=5,
                ticks=False,

            )
            )
)

bars = base.mark_bar().encode(
    color=alt.condition("datum.Growth > 0", alt.value(
        'rgb(136, 210, 214)'), alt.value('#D68C88')),
)

shade_df = pd.DataFrame([{'Date': '2019-04-01', 'Date2': '2025-04-01'}])

date_shading = alt.Chart(shade_df).mark_rect(
    color='#b4c8d8',
    opacity=0.25
).encode(
    x='Date:T',
    x2='Date2:T'
)

# add two lines of text: 2019-2025: 3.5% fall in Real Household Income per person
text_df_1 = pd.DataFrame([
    {'Date': '2019-04-01', 'Text': '2019-25', "Growth":0.08},
])
labels_1 = alt.Chart(text_df_1).mark_text(
    align='left',
    baseline='middle',
    font='Circular Std',
    color='#b4c8d8',
    dx=0,
    dy=0,
    fontSize=12,
    fontWeight=alt.expr("datum.Text == '2019-25' ? 'bold' : 'normal'"),
    opacity=1,
).encode(
    text='Text:N',
    x='Date:T',
    y=alt.value(-15),
)

# add two lines of text: 2019-2025: 3.5% fall in Real Household Income per person
text_df_2 = pd.DataFrame([
    {'Date': '2019-04-01', 'Text': '3.5% fall', "Growth":0.08},
])
labels_2 = alt.Chart(text_df_2).mark_text(
    align='left',
    baseline='middle',
    font='Circular Std',
    color='#b4c8d8',
    dx=0,
    dy=0,
    fontSize=10,
    fontWeight=alt.expr("datum.Text == '2019-25' ? 'bold' : 'normal'"),
    opacity=1,
).encode(
    text='Text:N',
    x='Date:T',
    y=alt.value(-5),
)

chart =labels_1 + labels_2 + date_shading + bars



chart = chart.properties(
    title={
        "text": ["The Cost of Living Squeeze"],
        "subtitle": ["Real Disposable Household Income per person, % Growth", "Source: OBR, 2022 and 2023"],
        "subtitleColor": "#b4c8d8",
        "fontSize": 12,
        "color": "#b4c8d8",
        "anchor": "start",
        "font": "Circular Std",
        "subtitleFont": "Circular Std",
        "dx": 30,
        "subtitleFontSize": 10,
    },
    width=300,
    height=200,
    background='#122b39',
).configure_view(
    strokeWidth=0
)
chart.save('disposable_household_income.png', scale_factor=4.0)
chart.save('disposable_household_income.json', indent=4)

chart
```
