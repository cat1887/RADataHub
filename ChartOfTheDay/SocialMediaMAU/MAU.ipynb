```python
import pandas as pd
import numpy as np
import eco_style
import altair as alt
alt.themes.enable("dark")
```

```python
df = pd.read_csv('social_media_platforms.csv')
df["Value"] = df["Value"]/1000
df = df.sort_values(by="Value", ascending=False)
df = df.head(10)
```

```python
base = alt.Chart(df).encode(
    y=alt.Y('Name:N', axis=alt.Axis(
        title='', format=""), sort='-x'),
    x=alt.X('Value:Q', axis=alt.Axis(labelExpr='datum.value + "B"')),
)

bars = base.mark_bar().encode(
    color=alt.condition(
        alt.datum.Name == 'Facebook',  # If the Name is Facebook,
        alt.value('#36B7B4'),     # color the bar 'lightblue'
        alt.value('#b4c8d8')   #  if not, color the bar 'steelblue'
    )
)

chart = bars

chart = chart.properties(
    title=alt.TitleParams(
        "Facebook: Three Billion Users",
        subtitle=["Social Media Platforms by Monthly Active Users, 2024", ""],
        anchor="start",
        fontSize=16,
        subtitleFontSize=12,
        color="whitesmoke",
        subtitleColor="#b4c8d8"
    )
)


chart.display()

chart.save("social_media_platforms.json")
chart.save("social_media_platforms.png", scale_factor=2.0)  
```

```python
fb_df = pd.read_csv("fb_mau.csv")

fb_df
# date format: Q3 '08
def convert_date(date):
    date = date.split(" '")
    year = date[1]
    quarter = int(date[0][-1])
    month = (quarter - 1) * 3 + 1
    date_str = f"20{year}-{month:02d}-01"
    return pd.to_datetime(date_str)

fb_df["date"] = fb_df["date"].apply(convert_date)
    
fb_df.to_csv("fb_mau_clean.csv", index=False)
```
