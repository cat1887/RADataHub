```python
# funding vs spending chart
income = pd.read_excel('income of HE providers.xlsx', skiprows=12)
expenses = pd.read_excel('expenditure of HE providers.xlsx', skiprows=11)
expenses['Total costs'] = expenses.iloc[:, 1:].sum(axis=1)
expenses['Other costs'] = expenses['Restructuring costs'] + expenses['Other operating expenses']
expenses['Non-operational costs'] = expenses['Depreciation and amortisation'] + expenses['Interest and other finance costs']

expenses
```

```python
expenses_negative = expenses[['Year']].join(expenses.iloc[:, 1:].mul(-1))
```

```python
income = income[income['Country'] == 'All'].reset_index(drop=True)
income['Grants and contracts'] = income['Funding body grants'] + income['Research grants and contracts']
income['Other Income'] = income['Other income'] + income['Investment income'] + income['Donations and endowments']
income
```

```python
joined_df = income.merge(expenses_negative, on='Year')
joined_df['Net Income'] = joined_df['Total costs'] + joined_df['Total income']
joined_df = joined_df.drop(columns=['Country', 'Total income', 'Total costs']) 
joined_df = joined_df.drop(columns=['Funding body grants', 'Research grants and contracts', 'Restructuring costs', "Other operating expenses",
                                    'Depreciation and amortisation','Interest and other finance costs','Other income','Investment income', 'Donations and endowments'])
joined_df
```

```python
data_long = joined_df.melt('Year', var_name='category', value_name='amount')


data_long['amount'] = data_long['amount'] * 1000000
data_long
```

```python
bar_data = data_long[data_long['category'] != 'Net Income']
line_data = data_long[data_long['category'] == 'Net Income']
```

```python
categories_order = bar_data['category'].unique()
categories_order
```

```python
# Assuming bar_data and line_data are already defined

# Define color palettes
cool_palette = ['#E6224B', '#FF5733', '#FF8D72']
warm_palette= ['#179FDB', '#0063AF', '#004080']

# Combine palettes for categories (assuming first few are warm, then cool)
combined_palette = warm_palette + cool_palette[:len(bar_data['category'].unique()) - len(warm_palette)]

# Create the stacked bar chart
bar_chart = alt.Chart(bar_data).mark_bar().encode(
    x='Year:O',
    y=alt.Y('amount:Q', stack='zero', axis=alt.Axis(
        title=None,
        labelExpr="""datum.value === 0 ? '0' :
                     datum.value > 0 ? ('£' + format(datum.value / 1e9, ',.0f') + ' B') :
                     ('-£' + format(-datum.value / 1e9, ',.0f') + ' B')"""
    )),
    color=alt.Color('category:N', scale=alt.Scale(range=combined_palette), sort=categories_order,
                     legend=None),
).properties(
    title='Income and Expenses'
)

rule = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='grey').encode(
    y='y:Q'
)

#line
line_chart = alt.Chart(line_data).mark_line(point=True).encode(
    x='Year:O',
    y='amount:Q',
    color=alt.value('black'),  
    tooltip=['amount','category', 'Year']
)

# Text label
text_label_1 = alt.Chart(pd.DataFrame({'x': ['2022/23'], 'y': [45000000000], 'text': ['Grants and contracts']})).mark_text(
    align='left',
    baseline='middle',
    fontSize=12,
    dx=25,
    color='#0063AF'
).encode(
    x=alt.X('x:O', axis=alt.Axis(title=None)),
    y=alt.Y('y:Q', axis=alt.Axis(title=None)),
    text='text:N'
)

text_label_2 = alt.Chart(pd.DataFrame({'x': ['2022/23'], 'y': [32000000000], 'text': ['Other income']})).mark_text(
    align='left',
    baseline='middle',
    fontSize=12,
    dx=25,
    color='#004080'
).encode(
    x=alt.X('x:O', axis=alt.Axis(title=None)),
    y=alt.Y('y:Q', axis=alt.Axis(title=None)),
    text='text:N'
)

text_label_3 = alt.Chart(pd.DataFrame({'x': ['2022/23'], 'y': [16000000000], 'text': ['Tuition fees and education contracts']})).mark_text(
    align='left',
    baseline='middle',
    fontSize=12,
    dx=25,
    color ='#179FDB'
).encode(
    x=alt.X('x:O', axis=alt.Axis(title=None)),
    y=alt.Y('y:Q', axis=alt.Axis(title=None)),
    text='text:N'
)

text_label_4 = alt.Chart(pd.DataFrame({'x': ['2022/23'], 'y': [-12000000000], 'text': ['Staff costs']})).mark_text(
    align='left',
    baseline='middle',
    fontSize=12,
    dx=30,
    color='#E6224B'
).encode(
    x=alt.X('x:O', axis=alt.Axis(title=None)),
    y=alt.Y('y:Q', axis=alt.Axis(title=None)),
    text='text:N'
)

text_label_5 = alt.Chart(pd.DataFrame({'x': ['2022/23'], 'y': [-32000000000], 'text': ['Other costs']})).mark_text(
    align='left',
    baseline='middle',
    fontSize=12,
    dx=25,
    color='#FF5733'
).encode(
    x=alt.X('x:O', axis=alt.Axis(title=None)),
    y=alt.Y('y:Q', axis=alt.Axis(title=None)),
    text='text:N'
)

text_label_6 = alt.Chart(pd.DataFrame({'x': ['2022/23'], 'y': [-46000000000], 'text': ['Non-operational costs']})).mark_text(
    align='left',
    baseline='middle',
    fontSize=12,
    dx=25, 
    color='#FF8D72'
).encode(
    x=alt.X('x:O', axis=alt.Axis(title=None)),
    y=alt.Y('y:Q', axis=alt.Axis(title=None)),
    text='text:N'
)

text_label_7 = alt.Chart(pd.DataFrame({'x': ['2022/23'], 'y': [0], 'text': ['Net income']})).mark_text(
    align='left',
    baseline='middle',
    fontSize=12,
    dx=25, 
    color='black'
).encode(
    x=alt.X('x:O', axis=alt.Axis(title=None)),
    y=alt.Y('y:Q', axis=alt.Axis(title=None)),
    text='text:N'
)

combined_chart = alt.layer(bar_chart, rule, line_chart, text_label_1, text_label_2, text_label_3,text_label_4, text_label_5,text_label_6, text_label_7).resolve_scale(
    y='shared'
)

combined_chart = combined_chart.properties(
    title={
        "text": ["Financial trends in British universities"],
        "subtitle": ["Expenditure and income sources of higher education providers", "Source: HESA", ""],
        "subtitleColor": "#676A86",
        "anchor": "start"
    }
)

combined_chart
```
