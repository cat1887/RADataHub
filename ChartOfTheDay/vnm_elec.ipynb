```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_json("http://api.economicsobservatory.com/vnm/elec?vega")
```

```python
df
```

```python
base = alt.Chart(df).encode(
    x=alt.X('date:T',
                        axis=alt.Axis(
                title=None,
                labelColor="#b4c8d8",
                labelFont="Circular Std",
                gridColor="#b4c8d8",
                gridDash=[1, 5],
                gridOpacity=0.5,
                labelAlign="center",
                domainOpacity=0.5,
                labelOpacity=1,
                labelPadding=5,
                ticks=False,

            )

            
            ),
    y=alt.Y('value:Q',
            axis=alt.Axis(
            title=None,
            gridColor="#b4c8d8",
            gridDash=[1, 5],
            labelFont="Circular Std",
            gridOpacity=0,
            labelExpr="datum.value == 0 ? '' : datum.value == 2500 ? datum.label + ' Kwh' : datum.label",
            
            tickOpacity=0,
            domainOpacity=0.5,
            labelColor="#b4c8d8",
            labelOpacity=1,
            labelPadding=5,
            domainColor='#b4c8d8',
            tickColor='#b4c8d8'),
            
            
            ),
    color=alt.Color('Country:N', legend=None, scale=alt.Scale(reverse=False, range=[   
         "#36B7B4",
    "#00A767",
])),
opacity=alt.Opacity('Sector:N', legend=None,  scale=alt.Scale(domain=["Financial", "Manufacturing", "Other"], range=[1, 0.7, 0.4, 0.8])),
)

bars = base.mark_line()

labels = base.transform_filter("year(datum.date) == 2021").mark_text(
    align='left',
    baseline='middle',
    font='Circular Std',
    color='#b4c8d8',
    limit=48,
    ellipsis=".",
    dx = 5,
    fontSize=10,
    text=alt.expr("format(datum.value, '.4r') + ' Kwh'"),
    opacity=1,
).encode(
    opacity=alt.value(1),
    color=alt.value('#b4c8d8'),
)

chart = bars + labels

chart = chart.properties(
    title={
        "text": ["Vietnamese Electricity Production"],
        "subtitle": ["Per capita Kilowatt-hours", "GSO through ECO API"],
                             "subtitleColor": "#b4c8d8",
        "fontSize": 12,
        "color": "#b4c8d8",
        "anchor": "start",
        "font": "Circular Std",
        "subtitleFont": "Circular Std",
        "dx": 60,
        "subtitleFontSize": 10,
    },
    width=300,
    height=200,
    background='#122b39',
).configure_view(
    strokeWidth=0
)

chart.save('vnm_elec.json')
chart.save('vnm_elec.png', scale_factor=3.0)
```
