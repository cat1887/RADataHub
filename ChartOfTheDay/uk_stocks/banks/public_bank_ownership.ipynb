```python
import pandas as pd
import numpy as np

import altair as alt
```

```python
df = pd.read_excel("public_sector_banks.xlsx", skiprows=3)
df.columns = ["Year", "Value"]
df = df.iloc[:27]
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

df["Value"] = df["Value"]/100
df = df.query("Year >= '1960-01-01'")

base = alt.Chart(df).encode(
    x=alt.X('Year:T',         axis=alt.Axis(
            title=None,
            gridColor="#b4c8d8",
            gridDash=[1, 5],
            labelFont="Circular Std",
            gridOpacity=0,
            domainOpacity=0.5,
            labelColor="#b4c8d8",
            labelOpacity=1,
            labelPadding=5,
            domainColor='#b4c8d8',
            tickColor='#b4c8d8'),),
    y=alt.Y('Value:Q',

            axis=alt.Axis(
                title=None,
                labelColor="#b4c8d8",
                labelFont="Circular Std",
                gridColor="#b4c8d8",
                gridDash=[1, 5],
                gridOpacity=0.5,
                domainOpacity=0.5,


                format="%",
                labelOpacity=1,
                labelFontSize=12,
                labelPadding=5,
                ticks=False,

            )
            )
)

bars = base.mark_area(color="rgb(136, 210, 214)", interpolate='step-after')

chart = bars

chart = chart.properties(
    title={
        "text": ["Government Share Ownership in the UK"],
        "subtitle": ["Public Sector ownership of UK shares (%)", "Source: ONS"],
        "subtitleColor": "#b4c8d8",
        "fontSize": 14,
        "color": "#b4c8d8",
        "anchor": "start",
        "font": "Circular Std",
        "subtitleFont": "Circular Std",
        "dx": 24,
        "subtitleFontSize": 12,
    },
    width=300,
    height=200,
    background='#122b39',
).configure_view(
    strokeWidth=0
)

chart.save('public_bank_ownership.png', scale_factor=3.0)
chart.save('public_bank_ownership.json')

chart
```
