```python
olymp = pd.read_excel('olympic_medals.xlsx')

olymp = olymp.melt(id_vars='Year', var_name='Country', value_name='Medals')
```

```python
olymp['Hosted'] = 'None'
olymp.loc[(olymp['Country'].isin(['Japan'])) & (olymp['Year'].isin([2020])), 'Hosted'] = 'Hosted'
olymp.loc[(olymp['Country'].isin(['China'])) & (olymp['Year'].isin([2008])), 'Hosted'] = 'Hosted'
olymp.loc[(olymp['Country'].isin([ 'Great Britain'])) & (olymp['Year'].isin([2012])), 'Hosted'] = 'Hosted'
olymp.loc[(olymp['Country'].isin(['Brazil'])) & (olymp['Year'].isin([2016])), 'Hosted'] = 'Hosted'

olymp
```

```python
base = alt.Chart(olymp).mark_bar().encode(
    x=alt.X('Year:O', title=None),
    y=alt.Y('Medals:Q', axis=alt.Axis(title='Medal Count')),
    color=alt.Color('Hosted:N',
        scale=alt.Scale(domain=['Hosted', 'None'], range=['#E6224B', '#122B39']),
        legend=alt.Legend(
            title=None,
            values=['Hosted'],
            orient='top'
        )
    ),
    tooltip=['Country', 'Medals', 'Year']
).properties(
    width=100,
    height=200
)
chart = base.facet(
    column=alt.Column('Country:N', header=alt.Header(title=None, labelAngle=0, labelOrient='bottom'), sort=['China', 'Great Britain', 'Brazil', 'Japan', 'France']),
    columns=5
)
chart = chart.properties(
        title={
        "text": ["Olympics: the home advantage"],
        "subtitle": ["Medal counts of the last 4 hosts", "Source: IOC", ""],
        "subtitleColor": "#676A86",
    }
)

chart
```
