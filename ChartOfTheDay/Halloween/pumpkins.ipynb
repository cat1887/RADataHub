```python
import pandas as pd
import altair as alt
```

```python
iso3_df = pd.read_csv("https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv")
```

```python
iso3_df
```

```python
df = pd.read_csv("pumpkin_wids.csv")
df = df[["ReporterISO3", "TradeValue in 1000 USD"]]
df.columns = ["iso3", "Value"]

df['Value'] = df['Value'].astype(float)*1000

df = df.merge(iso3_df[["alpha-3", "name"]], left_on="iso3", right_on="alpha-3", how="left")
# replace United States of America with USA
df.loc[df.name == "United States of America", "name"] = "USA"

df = df.sort_values(by="Value", ascending=False)
df = df[df.iso3 != "EUN"]
df = df.head(5)

"""
        "gridColor": "#b4c8d8",
        "format": "%",
        "gridDash": [1, 5],
        "gridOpacity": 0.5,
        "labelColor": "#b4c8d8",
        "labelOpacity": 0.7,
        "labelPadding": 5,
        "tickColor": "#b4c8d8",
        "tickCount": 8,
        "tickOpacity": 0.5,
        "ticks": false,
        "title": "Annual real US GDP growth | ONS through ECO API",
        "titleAlign": "left",
        "titleAngle": 0,
        "titleBaseline": "bottom",
        "titleColor": "#b4c8d8",
        "titleFontSize": 12,
        "titleOpacity": 0.8,
        "titleX": 0,
        "titleY": -7
"""


chart = alt.Chart(df.head(10)).mark_bar(color="rgb(234,121,42)").encode(
    x = alt.X("Value:Q",
        axis=alt.Axis(
            title=None,
            gridColor="#b4c8d8",
            gridDash=[1, 5],
            labelFont="Circular Std",
            gridOpacity=0.5,
            labelColor="#dbab5e",
            labelOpacity=1,
            labelPadding=5,
            domainColor= '#dbab5e',
            tickColor= '#dbab5e',
            format="$,.2s"),
     title="Trade Value in 1000 USD"),
    y = alt.Y("name:N", 
    axis=alt.Axis(
        title=None,
        labelColor="#dbab5e",
        labelFont="Circular Std",
        labelOpacity=1,
        labelFontSize=12,
        labelPadding=5,
        ticks=False,
    ),
    
    title="Country", sort="-x"),
)

"""
    "title": {
      "anchor": "start",
      "baseline": "top",
      "color": "#b4c8d8",
      "dx": 30,
      "fontSize": 14,
      "subtitleColor": "#000000",
      "subtitleFontSize": 12
    }
  },
"""


chart = chart.properties(
    title={
        "text": ["Top 5 Pumpkin Exporters"],
        "dx": 30,
        "fontSize": 16,
        "font": "Circular Std",
        "subtitleFont": "Circular Std",
        "color": "rgb(234,121,42)",
        "anchor": "start",
        "subtitleColor": "#b4c8d8",
        "subtitle": ["2022 Exports of 'Fresh or chilled pumpkins, squash and gourds' | World Bank "],
        "subtitleColor": "#dbab5e",
    },
    background='#122B39',
    width=500,
    height=400,
).configure_view(
    strokeOpacity=0
)

chart.display()

chart.save("pumpkins.png", scale_factor=2.0)
```
